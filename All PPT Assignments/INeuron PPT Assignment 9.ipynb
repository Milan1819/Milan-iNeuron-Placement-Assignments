Question no.1 :- Given an integer `n`, return *`true` if it is a power of two. Otherwise, return `false`*.

An integer `n` is a power of two, if there exists an integer `x` such that `n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

Answer no.1

```python
class Power:
    def is_power(self,n):

        if n == 1:
            return True
        if n < 1:
            return False
        return self.is_power(n/2)


obj = Power() 
n = 3
print(obj.is_power(n))
```

Question no.2 :- Given a number n, find the sum of the first natural numbers.

**Example 1:**

Input: n = 3 

Output: 6

**Example 2:**

Input  : 5 

Output : 15

Answer no.2

```python
class Natural_sum:
    def find_sum(self,n):

        if n <= 1:
            return n
        else:
            return n + self.find_sum(n-1)


obj1 = Natural_sum() 
n = 18
print(obj1.find_sum(n))
```

Question no.3 :- Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

Answer no.3

```python
class Factorial:
    def get_factorial(self,N):
        
        if N == 0 or N == 1:
            return 1
        elif N > 1:
            return N * self.get_factorial(N-1)


obj2 = Factorial() 
N = 5
print(obj2.get_factorial(N))
```

Question no.4 :- Given a number N and a power P, the task is to find the exponent of this number raised to the given power, i.e. N^P.

**Example 1 :** 

Input: N = 5, P = 2

Output: 25

**Example 2 :**
Input: N = 2, P = 5

Output: 32

Answer no.4

```python
class Power_p:
    def power(self,N,P):
        if P == 0:
            return 1
        return (N * self.power(N,P-1)) 

obj3 = Power_p()
N = 5
P = 2
print(obj3.power(N,P))
```

Question no.5 :- Given an array of integers **arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

Answer no.5

```python
class Largest:
    def is_largest(self,arr,n):

        if n == 1:
            return arr[0]
        return max(arr[n-1],self.is_largest(arr,n-1))


obj4 = Largest()
arr = [1,2,3,11,18,16]
n = len(arr) 
print(obj4.is_largest(arr,n))
```

Question no.6 :- Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

Answer no.6

```python
class Nth_term:
    def get_nth(self,a,d,N):
        return (a + (N-1) * d) 


obj5 = Nth_term()
a = 5
d = 2
N = 10
print(obj5.get_nth(a,d,N))
```

Question no.7 :- Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = “ABC”*

***Output:***

*“ABC”, “ACB”, “BAC”, “BCA”, “CBA”, “CAB”*

**Example 2:**

***Input:***

*S = “XY”*

***Output:***

*“XY”, “YX”*

Answer no.7

```python
import itertools
class Permutations:
    def all_pers(self,S):

        new_S = sorted(S)
        pers = itertools.permutations(new_S)
               
        return pers 


obj6 = Permutations()
S = "ABC"
all_p = obj6.all_pers(S) 
for i in all_p:
    print("".join(i),end = ",")
```

Question no.8 :- Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120

**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

Answer no.8

```python
class Product:
    def get_product(self,arr):
        p = 1
        for i in arr:
            p = p*i
        return p


obj7 = Product()
arr = {1,2,3,4,5}
print(obj7.get_product(arr))
```
