Question :- Implement a queue using a list in Python. Include the necessary methods such as enqueue, dequeue, and isEmpty.

Answer

```python
class Queue:

    def __init__(self):
        self.queue = []

    def isEmpty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if len(self.queue) < 1:
            return None
        return self.queue.pop(0)

    def display(self):
        print(self.queue)

    def size(self):
        return len(self.queue)


q = Queue()

q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)

print(q.isEmpty())

q.display()

q.dequeue()

print("After removing an element")
q.display()
```
