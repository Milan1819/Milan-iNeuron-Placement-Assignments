Question no.1 :- Given an integer array nums of 2n integers, group these integers into n pairs (a1,b1),(a2,b2),..., (an,bn) such that the sum of (ai,bi) for all i is maximized. Return the maximized sum. 

Example:

Input: nums = [1,4,3,2]

Output: 4

Explanation: All possible pairings (Ignoring the ordering of elements) are:

1.(1,4),(2,3) -> max(1,4) + max(2,3) = 4+3 = 7

2.(1,3),(2,4) -> max(1,3) + max(2,4) = 4+3 = 7

3.(1,2),(3,4) -> max(1,2) + max(3,4) = 4+2 = 6 

So the maximum possible sum is 7

Answer no.1

```python
class Max_sum_pair:
    def find_max_sum_pair(self,arr,n):

        maxsum = 0
        
        for i in range(0, n - 1):
            for j in range(i + 1, n):
                sum = arr[i] + arr[j]
                if (sum > maxsum):
                    maxsum = sum
        return maxsum

obj = Max_sum_pair() 
arr = [11,20,32,15]
n = len(arr) 
print(obj.find_max_sum_pair(arr,n))
```

Question no.2 :- Alice has n candies, where the ith candy is of the type candyType[i]. Alice noticed that she started to gain weight, so she visited a doctor. 

The doctor advised Alice to only eat n/2 of the candies she has(n is always even). Alice likes her candies very much, and she wants to eat the maximum number of different types of candies while still following the doctor's advice. 

Given the integer array candyType of length n, return the maximum number of different types of candies she can eat if she only eats n/2 of them. 

Example 1:

Imput: candyType = [1,1,2,2,3,3]

Output: 3

Explanation: Alice can only eat 6/2 candies. Since there are only 3 types, she can eat one of each type.

Answer no.2

```python
class Alice_candies:
    def distribute_candies(self,candytype):

        type_of_candies = len(set(candytype))

        return min(type_of_candies,len(candytype)//2) 

obj1 = Alice_candies() 
candytype = [1,1,2,2,3,3,4,4]
print(obj1.distribute_candies(candytype))
```

Question no.3 :- We define a harmonious array as an array where the difference between its maximum value and its minimum value is exactly 1.

Given an integer array nums, return the length of its longest harmonious subsequence among all the possible subsequences. 

A subsequence of an array is a sequence that can be derived from the array by deleting some or no elements without changing the order of the remaining elements. 

Example 1:

Input: nums = [1,3,2,2,5,2,3,7]

Output: 5

Explanation: The Longest harmonious subsequence is [3,2,2,2,3].

Answer no.3

```python
from collections import Counter

class Harmonious_subsq:
    def longest_har_subsq(self,nums):
        
        count=Counter(nums)
        result=0
        
        for ele in count:
            if count[ele+1]:
                result = max(result, count[ele]+count[ele+1])
        
        return result

obj2 = Harmonious_subsq() 
nums = [1,3,2,2,5,2,3,7]
print(obj2.longest_har_subsq(nums))
```

Question no.4 :- You have a long flowerbed in which some of the plots are planted, and some are not. However, flowers cannot be planted in adjacent plots. 

Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty, and an integer n, return true if n new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false otherwise. 

Example 1:

Input: flowerbed = [1,0,0,0,1],n = 1

Output: True

Answer no.4

```python
class Place_flowerbed:
    def can_place_flowers(self,flowerbed,n):

        l=len(flowerbed)
        flowerbed=[0]+flowerbed+[0]

        for i in range(1,l-1):

            if flowerbed[i]==0 and flowerbed[i-1]==0 and flowerbed[i+1]==0:
                flowerbed[i]=1
                n-=1

        if n<=0:
            return True
        return False

obj3 = Place_flowerbed()
flowerbed = [1,0,0,0,1]
n = 1
print(obj3.can_place_flowers(flowerbed,n))
```

Question no.5 :- Given an integer array nums, find the numbers whose product is maximum and return the maximum product. 

Example 1:

Input: nums = [1,2,3]

Output : 6

Answer no.5

```python
class Product:
    def max_product(self,list):

        max_prod = 1
        for i in list:
            max_prod = max_prod*i
        
        return max_prod

obj4 = Product() 
list = [1,2,3,4,5]
print(obj4.max_product(list))
```

Question no.6 :- Given an array of integers nums which is sorted in ascending order, and an integer target, write a function to search target in nums. If target exists, then return its index. Otherwise, return -1. 

You must write an algorithm with O(log n) runtime complexity. 

Example 1:

Input: nums = [-1,0,3,5,9,12], target = 9

Output: 4

Explanation: 9 exists in nums and its index is 4.

Answer no.6

```python
class Target:
    def find_target(self,L,target):

        for i in L:
            if i == target:
                return L.index(i) 
        return -1
            
            

obj5 = Target() 
L = [1,3,6,17,18,24,25,38]
target = 18
print(obj5.find_target(L,target))
```

Question no.7 :- An array is monotonic if it either monotone increasing or monotone decreasing. 

An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is monotone decreasing if for all i <= j, nums[i] <= nums[j].

Given an integer array nums, return true if the given array is monotonic, or false otherwise. 

Example 1:

Input: nums = [3,2,2,3]

Output: True

Answer no.7

```python
class Monotonic:
    def is_monotonic(self,array):
        
        if array[::] == array[::-1]:
            return True 
        return False

obj6 = Monotonic() 
array = [70,2,40,2,70]
print(obj6.is_monotonic(array))
```

Question no.8 :- You are given an integer array nums and an integer k.

In one operation, you can choose any index i where 0 <= i nums.length and change nums[i] to nums[i]+x where x is an integer from the range [-k,k]. You can apply this operation at most once for each index i.

The score of nums is the difference between the maximum and minimum elements in nums. 

Return the minimum score of sums after applying the mentioned operation at most once for each index in it. 

Example 1:

Input: nums = [1], k = 0

Output: 0

Explanation: The score is max(nums) - min(nums) = 1-1 = 0.

Answer no.8

```python
class Difference:
    def get_score(self,N,k):
       
        Max=max(N)
        Min=min(N)

        return max(0,(Max-k)-(Min+k))

obj7 = Difference() 
N = [1,2,3,4,5,6,8]
k = 0
print(obj7.get_score(N,k))
```
