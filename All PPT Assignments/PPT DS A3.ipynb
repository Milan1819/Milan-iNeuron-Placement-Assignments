Question no.1 :- A company wants to analyze the sales performance of its products in different regions. They have collected the following data:

   Region A: [10, 15, 12, 8, 14]

   Region B: [18, 20, 16, 22, 25]

   Calculate the mean sales for each region.

Answer no.1

```python
import statistics

Region_A = [10, 15, 12, 8, 14]
Region_B = [18, 20, 16, 22, 25]

# region-vise mean sales
mean_A = statistics.mean(Region_A)
mean_B = statistics.mean(Region_B)

print("Mean Sales for Region A is:",mean_A)
print("Mean Sales for Region B is:",mean_B)
```

Question no.5 :- A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:

   Group A: [85, 90, 92, 88, 91]

   Group B: [82, 88, 90, 86, 87]

   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.

Answer no.5

```python
from scipy import stats

Group_A = [85, 90, 92, 88, 91]
Group_B = [82, 88, 90, 86, 87]

# take significance_value
Significance_value = 0.05

# performing t-test
statistic,pvalue = stats.ttest_ind(Group_A,Group_B)

# p-value
P_value = pvalue

# deciding hypothesis based on significance-value and p-value
if P_value > Significance_value:
    print("There is no significant difference in the mean scores between two groups")
else:
    print("There is significant difference in the mean scores between two groups")
```

Question no.9 :- A study is conducted to compare the effectiveness of two different medications. The recovery times of the patients in each group are as follows:

   Medication A: [10, 12, 14, 11, 13]

   Medication B: [15, 17, 16, 14, 18]

   Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications.

Answer no.9

```python
Med_A = [10, 12, 14, 11, 13]
Med_B = [15, 17, 16, 14, 18]

# take significance value
Significance_value = 0.05

# perform Annova test
Annova_test = stats.f_oneway(Med_A,Med_B)

# p-value
P_value = Annova_test.pvalue

# deciding hypothesis based on Annova test
if P_value > Significance_value:
    print("There is no significant difference in the mean recovery times between two medications")
else:
    print("There is significant difference in the mean recovery times between two medications")
```

Question no.11 :- A quality control department wants to test the weight consistency of a product. The weights of a sample of products are as follows:

    [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]

    Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams.

Answer no.11

```python
products = [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]

mean = 10

# take significance value
Significance_value = 0.05

# perform t-test for one sample mean
T_test_mw = stats.ttest_1samp(products,popmean = mean)

# p-value
P_value = T_test_mw.pvalue

# deciding hypothesis on based of one sample ttest
if P_value > Significance_value:
    print("Mean weights do not differ significantly from 10 grams")
else:
    print("Mean weights differ significantly from 10 grams")
```

Question no.12 :- A company wants to analyze the click-through rates of two different website designs. The number of clicks for each design is as follows:

    Design A: [100, 120, 110, 90, 95]

    Design B: [80, 85, 90, 95, 100]

    Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.

Answer no.12

```python
Design_A = [100, 120, 110, 90, 95]

Design_B = [80, 85, 90, 95, 100]

# take significance value
Significance_value = 0.05

# perform the chi2 test
Chi_test = stats.chi2_contingency(Design_A,Design_B)

# p-value
P_value = Chi_test.pvalue

# deciding hypothesis based on chi2 test
if P_value > Significance_value:
    print("There is no significance difference in the click-through rates between the two designs")
else:
    print("There is significance difference in the click-through rates between the two designs")
```

Question no.13 :- A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:

    [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

    Calculate the 95% confidence interval for the population mean satisfaction score.

Answer no.13

```python
cus_s = [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

# creating confidence interval of 95%
CI = stats.t.interval(confidence=0.95, df=len(cus_s)-1, loc=np.mean(cus_s), scale=stats.sem(cus_s))
print("The 95% Confidence Interval for population mean satisfaction score is:",CI)
```

Question no.15 :- A study is conducted to compare the preferences of two groups of participants. The preferences are measured on a Like scale from 1 to 5. The data collected is as follows:

    Group A: [4, 3, 5, 2, 4]

    Group B: [3, 2, 4, 3, 3]

    Perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between the two groups.

Answer no.15

```python
Group_A = [4, 3, 5, 2, 4]
Group_B = [3, 2, 4, 3, 3]

# take significance value
Significance_value = 0.05

# perform Mann-Whitney U test
MWU = stats.mannwhitneyu(Group_A,Group_B)

# p-value
P_value = MWU.pvalue

# deciding hypothesis based on Mann-Whitney U test
if P_value > Significance_value:
    print("There is no significance difference in the median preferences between the two groups")
else:
    print("There is significance difference in the median preferences between the two groups")
```

Question no.17 :- A study is conducted to compare the performance of three different machine learning algorithms. The accuracy scores for each algorithm are as follows:

    Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]

    Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]

    Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]

    Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.

Answer no.17

```python
Algorithm_A = [0.85, 0.80, 0.82, 0.87, 0.83]
Algorithm_B = [0.78, 0.82, 0.84, 0.80, 0.79]
Algorithm_C = [0.90, 0.88, 0.89, 0.86, 0.87]

# take significance value
Significance_value = 0.05

# perform Kruskal-Wallis test
KWT = stats.mstats.kruskalwallis(Algorithm_A,Algorithm_B,Algorithm_C)

# p-value
P_value = KWT.pvalue

# deciding hypothesis based on Kruskal-Wallis test
if P_value > Significance_value:
    print("There is no significance difference in the median accuracy scores between the algorithms")
else:
    print("There is significance difference in the median accuracy scores between the algorithms")
```
