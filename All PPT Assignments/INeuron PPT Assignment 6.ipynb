Question no.1 :- A permutation perm of n + 1 integers of all the integers in the range [0, n] can be represented as a string s of length n where:

- s[i] == 'I' if perm[i] < perm[i + 1], and
- s[i] == 'D' if perm[i] > perm[i + 1].

Given a string s, reconstruct the permutation perm and return it. If there are multiple valid permutations perm, return **any of them**.

**Example 1:**

**Input:** s = "IDID"

**Output:**

[0,4,1,3,2]

Answer no.1

```python
class Str_permutations:
    def string_match(self,s):
        i, n = 0, len(s)
        ans = []
        for c in s:
            if c == 'I': 
                ans.append(i)
                i+=1
            else:
                ans.append(n)
                n-=1
        ans.append(i)
        return ans

obj = Str_permutations() 
s = "IDID"
print(obj.string_match(s))
```

Question no.2 :- You are given an m x n integer matrix with the following two properties:

- Each row is sorted in non-decreasing order.
- The first integer of each row is greater than the last integer of the previous row.

Given an integer target, return true *if* target *is in* matrix *or* false *otherwise*.

You must write a solution in O(log(m * n)) time complexity.

Example 1:

**Input:** matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3

**Output:** true

Answer no.2

```python
class Matrix_target:
    def get_target(self,matrix,target):
       
        for ele in matrix:
            if ele[-1] >= target:
               return target in ele
        return False  
                        

obj1 = Matrix_target() 
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 100
print(obj1.get_target(matrix,target))
```

Question no.3 :- Given an array of integers arr, return *true if and only if it is a valid mountain array*.

Recall that arr is a mountain array if and only if:

- arr.length >= 3
- There exists some i with 0 < i < arr.length - 1 such that:
- arr[0] < arr[1] < ... < arr[i - 1] < arr[i]
- arr[i] > arr[i + 1] > ... > arr[arr.length - 1]

**Example 1:**

**Input:** arr = [2,1]

**Output:**

false

Answer no.3

```python
class Mountain_array:
    def get_array(self,arr): 

        upFlag = True
        downFlag = False
    
        if len(arr) < 3:
            return False
    
        for i in range(1,len(arr)):
            if arr[i] > arr[i-1] and upFlag:
                downFlag = True            
            elif arr[i] < arr[i-1] and downFlag:
                upFlag = False 
            else:
                return False
        return not upFlag and downFlag

obj2 = Mountain_array() 
arr = [1,2,3,5,4,0]
print(obj2.get_array(arr))
```

Question no.4 :- Given a binary array nums, return *the maximum length of a contiguous subarray with an equal number of* 0 *and* 1.

**Example 1:**

**Input:** nums = [0,1]

**Output:** 2

**Explanation:**

[0, 1] is the longest contiguous subarray with an equal number of 0 and 1.

Answer no.4

```python
class Contiguous:
    def max_ones_zeros(self,arr):
        
        n=len(arr)
        for i in range(n):
            if arr[i]==0:
                arr[i]=-1
           
        maxLen=0
        sum=0
        d={}
        
        for i in range(n):
            sum = sum + arr[i]
            
            if sum==0:
                maxLen = i+1
                
            if sum in d:
                maxLen = max(maxLen, i-d[sum])
                 
            if sum not in d:
                d[sum]=i
                
        return maxLen

obj3 = Contiguous() 
arr = [0,1,0,1,1]
print(obj3.max_ones_zeros(arr))
```

Question no.5 :- The **product sum** of two equal-length arrays a and b is equal to the sum of a[i] * b[i] for all 0 <= i < a.length (**0-indexed**).

- For example, if a = [1,2,3,4] and b = [5,2,3,1], the **product sum** would be 1*5 + 2*2 + 3*3 + 4*1 = 22.

Given two arrays nums1 and nums2 of length n, return *the **minimum product sum** if you are allowed to **rearrange** the **order** of the elements in* nums1.

**Example 1:**

**Input:** nums1 = [5,3,4,2], nums2 = [4,2,2,5]

**Output:** 40

**Explanation:**

We can rearrange nums1 to become [3,5,4,2]. The product sum of [3,5,4,2] and [4,2,2,5] is 3*4 + 5*2 + 4*2 + 2*5 = 40.

Answer no.5

```python
class Product_sum:
    def min_mul_sum(self,nums1,nums2):
        nums1.sort()
        nums2.sort(reverse=True)
        ans = 0
        for i in range(len(nums1)):
            ans += nums1[i] * nums2[i]
        return ans

obj4 = Product_sum() 
nums1 = [5,3,4,2]
nums2 = [4,2,2,5]
print(obj4.min_mul_sum(nums1,nums2))
```

Question no.6 :- An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

Answer no.6

```python
from collections import deque
class Old_Arr:
    def get_old_arr(self,array):

        array.sort()
        que=deque([])
        output=[]
        
        for i in array:
            if que and que[0]==i:
                que.popleft()
            else:
                que.append(i*2)
                output.append(i)
        if que: return []
        return output


obj5 = Old_Arr() 
array = [1,3,4,2,6,8]
print(obj5.get_old_arr(array))
```

Question no.7 :- Given a positive integer n, generate an n x n matrix filled with elements from 1 to n2 in spiral order.

**Example 1:**

**Input:** n = 3

**Output:** [[1,2,3],[8,9,4],[7,6,5]]

Answer no.7

```python
from itertools import cycle
class Spiral_matrix:
    def get_spiral(self,n):

        matrix = [[0] * n for _ in range(n)]
        directions = cycle([(0, 1), (1, 0), (0, -1), (-1, 0)])
        row, col = 0, 0
        x, y = next(directions)

        for i in range(1, n ** 2 + 1):
            matrix[row][col] = i
            if not (0 <= row + x < n and 0 <= col + y < n and matrix[row + x][col + y] == 0):
                x, y = next(directions)
            row += x
            col += y  
        return matrix

obj6 = Spiral_matrix()
n = 3
print(obj6.get_spiral(n))
```

Question no.8 :- Given two [sparse matrices](https://en.wikipedia.org/wiki/Sparse_matrix) mat1 of size m x k and mat2 of size k x n, return the result of mat1 x mat2. You may assume that multiplication is always possible.

**Example 1:**

**Input:** mat1 = [[1,0,0],[-1,0,3]], mat2 = [[7,0,0],[0,0,0],[0,0,1]]

**Output:**

[[7,0,0],[-7,0,3]]

Answer no.8

```python
class Sparse_mul:
    def multiply_sparse(self,mat1,mat2):
        r1, c1, c2 = len(mat1), len(mat1[0]), len(mat2[0])
        res = [[0] * c2 for _ in range(r1)]
        for i in range(r1):
            for j in range(c2):
                for k in range(c1):
                    res[i][j] += mat1[i][k] * mat2[k][j]
        return res

obj7 = Sparse_mul() 
mat1 = [[1,0,0],[-1,0,3]]
mat2 = [[7,0,0],[0,0,0],[0,0,1]]
print(obj7.multiply_sparse(mat1,mat2))
```
