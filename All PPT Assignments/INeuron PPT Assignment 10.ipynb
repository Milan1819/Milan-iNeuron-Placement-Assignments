Question no.1 :- Given an integer `n`, return *`true` if it is a power of three. Otherwise, return `false`*.

An integer `n` is a power of three, if there exists an integer `x` such that `n == 3x`.

**Example 1:**


Input: n = 27
Output: true

Explanation: 27 = 3**3


**Example 2:**


Input: n = 0
Output: false

Explanation: There is no x where 3x = 0.


**Example 3:**


Input: n = -1
Output: false

Explanation: There is no x where 3x = (-1).

Answer no.1

```python
class Power_three:
    def is_power(self,n):
        if n==1 or n==3:
            return True
        if n>3 and n%3==0:
            return self.is_power(n/3)
        return False


obj = Power_three() 
n = 7
print(obj.is_power(n))
```

Question no.2 :- You have a list `arr` of all integers in the range `[1, n]` sorted in a strictly increasing order. Apply the following algorithm on `arr`:

- Starting from left to right, remove the first number and every other number afterward until you reach the end of the list.
- Repeat the previous step again, but this time from right to left, remove the rightmost number and every other number from the remaining numbers.
- Keep repeating the steps again, alternating left to right and right to left, until a single number remains.

Given the integer `n`, return *the last number that remains in* `arr`.

**Example 1:**


Input: n = 9

Output: 6

Explanation:
arr = [1, 2,3, 4,5, 6,7, 8,9]
arr = [2,4, 6,8]
arr = [2, 6]
arr = [6]

**Example 2:**

Input: n = 1

Output: 1

Answer no.2

```python

class Last_remaining:
    def last_one(self,n):
            if(n == 1):
                return 1        
            return 2 * ((n // 2 - self.last_one(n // 2)) + 1)


obj1 = Last_remaining()
n = 9
print(obj1.last_one(n))
```

Question no.3 :- Given a set represented as a string, write a recursive code to print all subsets of it. The subsets can be printed in any order.

**Example 1:**

Input :  set = “abc”

Output : { “”, “a”, “b”, “c”, “ab”, “ac”, “bc”, “abc”}

**Example 2:**

Input : set = “abcd”

Output : { “”, “a” ,”ab” ,”abc” ,”abcd”, “abd” ,”ac” ,”acd”, “ad” ,”b”, “bc” ,”bcd” ,”bd” ,”c” ,”cd” ,”d” }

Answer no.3

```python
class Subsets:
    def get_subsets(self,nums):
        res, n = [], len(nums)

        def sub(i, l):
            if i == n:
                res.append(l)
                return
            sub(i + 1, l + [nums[i]])
            sub(i + 1, l)
        sub(0,[])
        return res 


obj2 = Subsets() 
nums = "abc"
print(obj2.get_subsets(nums))
```

Question no.4 :- Given a string calculate length of the string using recursion.

**Examples:**


Input : str = "abcd"

Output :4

Input : str = "GEEKSFORGEEKS"

Output :13

Answer no.4

```python
class Length_recur:
    def get_len(self,str):
        if str == '':
            return 0
        else :
            return 1 + self.get_len(str[1:])
     
obj3 = Length_recur() 
str = "BATMANVSSUPERMAN"
print(obj3.get_len(str))
```

Question no.5 :- We are given a string S, we need to find count of all contiguous substrings starting and ending with same character.

**Examples :**

Input  : S = "abcab"

Output : 7

There are 15 substrings of "abcab"
a, ab, abc, abca, abcab, b, bc, bca
bcab, c, ca, cab, a, ab, b
Out of the above substrings, there
are 7 substrings : a, abca, b, bcab,
c, a and b.

Input  : S = "aba"

Output : 4

The substrings are a, b, a and aba.

Answer no.5

```python
class Contiguous:
    def get_contiguous(self,S):
        result = 0
        n = len(S)

        for i in range(n):
            for j in range(i, n):
                if (S[i] == S[j]):
                    result = result + 1
        return result


obj4 = Contiguous()
S = "abcab"
print(obj4.get_contiguous(S))
```

Question no.6 :- The tower of Hanoi is a famous puzzle where we have three rods and N disks. The objective of the puzzle is to move the entire stack to another rod. Also, you need to find the total moves.Note: The discs are arranged such that the top disc is numbered 1 and the bottom-most disc is numbered N. **Example 1:**

Input:

move disk 1 from rod 1 to rod 2

move disk 2 from rod 1 to rod 3

move disk 1 from rod 2 to rod 3

Explanation: For N=2 , steps will be
as follows in the example and total
3 steps will be taken.

Answer no.6

```python
def Tower_of_Hanoi(n,source,destination,auxiliary):
    if n==1:
        print ("Move disk 1 from source",source,"to destination",destination)
        return
    Tower_of_Hanoi(n-1, source, auxiliary, destination)
    print ("Move disk",n,"from source",source,"to destination",destination)
    Tower_of_Hanoi(n-1, auxiliary, destination, source)
         

n = 2
TowerOfHanoi(n,'1','2','3')
```

Question no.7 :- Given a string **str**, the task is to print all the permutations of **str**. A **permutation** is an arrangement of all or part of a set of objects, with regard to the order of the arrangement. For instance, the words ‘bat’ and ‘tab’ represents two distinct permutation (or arrangements) of a similar three letter word.

**Examples:**

Input: str = “cd”

**Output:** cd dc

**Input:** str = “abb”
 
**Output:** abb abb bab bba bab bba

Answer no.7

```python
import itertools
class Permutations:
    def all_pers(self,S):

        new_S = sorted(S)
        pers = itertools.permutations(new_S)
               
        return pers 


obj6 = Permutations()
S = "abb"
all_p = obj6.all_pers(S) 
for i in all_p:
    print("".join(i),end = " ")
```

Question no.8 :- Given a string, count total number of consonants in it. A consonant is an English alphabet character that is not vowel (a, e, i, o and u). Examples of constants are b, c, d, f, and g.

**Examples :**

Input: "abc de"

Output: 3

There are 3 consonants b, c and d.

Input: "geeksforgeeks portal"

Output: 12

Answer no.8

```python
class Consonants:
    def count_them(self,S):
        consonants = 0
        vowels = 0

        for i in S:
            if i == 'a'or i == 'e'or i == 'i'or i == 'o'or i == 'u':
                vowels = vowels+1;
            else:
                consonants = consonants+1;                         
        
        return consonants


obj7 = Consonants()
S = "geeksforgeeksportal"
print(obj7.count_them(S))
```
