Question no.8 :- Implement a function to check if a given number is prime.

Answer no.8

```python
def is_Prime(n1):

    if n1 < 0:
        return "You entered negative number"

    elif n1 == 1:
        return "1 is neither prime nor composite"

    else:
        for i in range(2,n1):
            if n1 % i == 0:
                return False
        return True


n1 = 41
is_Prime(n1)
```

Question no.18 :- Implement a function to find the maximum subarray sum in a given list.

Answer no.18

```python
def max_sum_of_subarray(array,n,k):
    max_sum = 0

    for i in range(0,n-k+1):
        temp = 0
        for j in range(i,i+k):
            temp += array[j]

        if (temp > max_sum):
            max_sum = temp

    return max_sum


array = [1,2,3,4,5,7,8,9]
n = len(array)
k = 3
max_sum_of_subarray(array,n,k)
```

Question no.34 :- Implement a function to find the longest common prefix among a list of strings.

Answer no.34

```python

def longest_common_prefix(List4):

    pre = List4[0]

    for i in List4:
        while not i.startswith(pre):
            pre = pre[:-1]

    return pre


List4 = ["Dog","Door","Doomsday","Doctor"]
longest_common_prefix(List4)
```
