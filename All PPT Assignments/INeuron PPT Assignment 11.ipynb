Question no.1 :- Given a non-negative integer `x`, return *the square root of* `x` *rounded down to the nearest integer*. The returned integer should be **non-negative** as well.

You **must not use** any built-in exponent function or operator.

- For example, do not use `pow(x, 0.5)` in c++ or `x ** 0.5` in python.

**Example 1:**

Input: x = 4

Output: 2

Explanation: The square root of 4 is 2, so we return 2.

Answer no.1

```python
class Square_root:
    def sqrt(self,x):

        low = 0
        high = x

        while low <= high:
            mid = low + (high - low) // 2
            if mid * mid == x:
                return mid
            elif mid * mid < x:
                ans = mid
                low = mid + 1
            else:
                high = mid - 1
        return ans


obj = Square_root()
x = 16
print(obj.sqrt(x))
```

Question no.2 :- A peak element is an element that is strictly greater than its neighbors.

Given a **0-indexed** integer array `nums`, find a peak element, and return its index. If the array contains multiple peaks, return the index to **any of the peaks**.

You may imagine that `nums[-1] = nums[n] = -âˆž`. In other words, an element is always considered to be strictly greater than a neighbor that is outside the array.

You must write an algorithm that runs in `O(log n)` time.

**Example 1:**

Input: nums = [1,2,3,1]

Output: 2

Explanation: 3 is a peak element and your function should return the index number 2.

Answer no.2

```python
class Peak_ele:
    def get_peak(self,nums):

        left = 0
        right = len(nums)-1
        
        while left < right:
            mid = (left + right) // 2
            if nums[mid] > nums[mid+1]:
                right = mid
            else:
                left = mid + 1                
        return left


obj1 = Peak_ele() 
nums = [1,2,3,1]
print(obj1.get_peak(nums))
```

Question no.3 :- Given an array `nums` containing `n` distinct numbers in the range `[0, n]`, return *the only number in the range that is missing from the array.*

**Example 1:**

Input: nums = [3,0,1]

Output: 2

Explanation: n = 3 since there are 3 numbers, so all numbers are in the range [0,3]. 2 is the missing number in the range since it does not appear in nums.

Answer no.3

```python
class Lost:
    def missing(self,nums):

        low = 0 
        high = len(nums)-1
        nums.sort()

        while low <= high:
            mid = (low+high)//2
            if mid == nums[mid]:
                low = mid+1
            elif mid != nums[mid]:
                high = mid-1
        return high+1
obj2 = Lost() 
nums = [3,0,1]
print(obj2.missing(nums))
```

Question no.4 :- Given an array of integers `nums` containing `n + 1` integers where each integer is in the range `[1, n]` inclusive.

There is only **one repeated number** in `nums`, return *this repeated number*.

You must solve the problem **without** modifying the array `nums` and uses only constant extra space.

**Example 1:**

Input: nums = [1,3,4,2,2]

Output: 2

Answer no.4

```python
class Duplicate:
    def get_duplicate(self,nums):
        nums.sort()

        for i in range(len(nums)):
            l, r = i + 1, len(nums) - 1
            while l <= r:
                mid = l + ((r - l) // 2)
                if nums[i] == nums[mid]:
                    return nums[mid]
                elif nums[i] <= nums[mid]:
                    r = mid - 1
                else:
                    l = mid + 1
        return nums[mid]


obj3 = Duplicate() 
nums = [1,3,4,2,2]
print(obj3.get_duplicate(nums))
```

Question no.5 :- Given two integer arrays `nums1` and `nums2`, return *an array of their intersection*. Each element in the result must be **unique** and you may return the result in **any order**.

**Example 1:**

Input: nums1 = [1,2,2,1], nums2 = [2,2]

Output: [2]

Answer no.5

```python
class Intersection:
    def get_common(self,nums1,nums2):
        nums1 = list(set(nums1))
        nums2 = list(set(nums2))
        arr = []
        for i in nums1:
            if i in nums2:
                arr.append(i)                
        return arr


obj4 = Intersection()
nums1 = [1,2,4,5,7]
nums2 = [2,3,4,8,9]
print(obj4.get_common(nums1,nums2))
```

Question no.6 :- Suppose an array of length `n` sorted in ascending order is **rotated** between `1` and `n` times. For example, the array `nums = [0,1,2,4,5,6,7]` might become:

- `[4,5,6,7,0,1,2]` if it was rotated `4` times.
- `[0,1,2,4,5,6,7]` if it was rotated `7` times.

Notice that **rotating** an array `[a[0], a[1], a[2], ..., a[n-1]]` 1 time results in the array `[a[n-1], a[0], a[1], a[2], ..., a[n-2]]`.

Given the sorted rotated array `nums` of **unique** elements, return *the minimum element of this array*.

You must write an algorithm that runs in `O(log n) time.`

**Example 1:**

Input: nums = [3,4,5,1,2]

Output: 1

Explanation: The original array was [1,2,3,4,5] rotated 3 times.

Answer no.6

```python
class Minimum:
    def get_min(self,nums):
        low, high = 0, len(nums) - 1

        while low < high:
            mid = low + (high - low) // 2
            if nums[mid] > nums[high]:
                low = mid + 1
            elif nums[mid] < nums[high]:
                high = mid
        return nums[low]


obj5 = Minimum()
nums = [3,4,5,1,2]
print(obj5.get_min(nums))
```

Question no.7 :- Given an array of integers `nums` sorted in non-decreasing order, find the starting and ending position of a given `target` value.

If `target` is not found in the array, return `[-1, -1]`.

You must write an algorithm with `O(log n)` runtime complexity.

**Example 1:**

Input: nums = [5,7,7,8,8,10], target = 8

Output: [3,4]

Answer no.7

```python
class Target:
    def get_target(self,nums,target):
        start = 0
        end = len(nums)-1

        while start <= end:
            mid = (start+end) // 2
            if nums[start] == nums[end] == target:
                return [start, end]
            if nums[mid] < target:
                start = mid+1
            elif nums[mid] > target:
                end = mid-1
            else:
                if nums[start] != target:
                    start += 1
                if nums[end] != target: 
                    end -= 1
        return [-1,-1]
   

obj6 = Target()
nums = [5,7,7,8,8,10]
target = 8
print(obj6.get_target(nums,target))
```

Question no.8 :- Given two integer arrays `nums1` and `nums2`, return *an array of their intersection*. Each element in the result must appear as many times as it shows in both arrays and you may return the result in **any order**.

**Example 1:**

Input: nums1 = [1,2,2,1], nums2 = [2,2]

Output: [2,2]

Answer no.8

```python
from collections import Counter

class Intersection2:
    def get_intersect(self,nums1,nums2):
        d1 = Counter(nums1)
        d2 = Counter(nums2)
        d12 = (d1 & d2)

        res = []
        for item in d12.items():
            res.extend([item[0]]*item[1])
        return res


obj7 = Intersection2() 
nums1 = [1,2,2,1]
nums2 = [2,2]
print(obj7.get_intersect(nums1,nums2))
```
