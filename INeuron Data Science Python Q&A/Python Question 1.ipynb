Question no.1 :- Write a program that takes a string as input, and counts the frequency of each word in the string, there might be repeated characters in the string. Your task is to find the highest frequency and returns the length of the highest-frequency word.

Answer no.1

```python
from collections import Counter
class High_freq_length:
    def find_highest_freq_length(self,string):
        try:
            # Split the string 
            words = string.split()

            # Count frequency of each word
            word_counts = Counter(words)

            # Find the word/s with the highest frequency
            max_frequency = max(word_counts.values())
            highest_frequency_words = [word for word, count in word_counts.items() if count == max_frequency]

            # Find length of highest frequency word
            highest_frequency_length = max(len(word) for word in highest_frequency_words)

            return highest_frequency_length
        except Exception as e:
            print(" Error occurred:", str(e))
            return None


obj = High_freq_length() 
string = "I will support support support RCB RCB RCB for entire life"
output = obj.find_highest_freq_length(string)
print(f"The length of highest frequency word is: {output}")
```
