Question no.2 :- Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if he can remove just one character at the index in the string, and the remaining characters will occur the same number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO.

Answer no.2

```python
from collections import Counter
class String_validation:
    def get_str_validity(self,S):
        try:
            # Count frequency of each character in the string
            char_counts = Counter(S)

            # Count total of frequencies
            frequency_counts = Counter(char_counts.values())

            # If all frequencies are the same
            if len(frequency_counts) == 1 or (len(frequency_counts) == 2 and frequency_counts[1] == 1):
                return "YES"
            else:
                return "NO"
        except Exception as e:
            print("Error occurred:", str(e))
            return "NO"


obj = String_validation() 
S = "aabbdd336600"
output = obj.get_str_validity(S)
print(output)
```
