```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize


def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Perform part-of-speech tagging
    pos_tags = pos_tag(words)

    # Initialize count variables
    verb_count = 0
    noun_count = 0
    pronoun_count = 0
    adjective_count = 0

    # Count the number of each part-of-speech
    for word, pos in pos_tags:
        if pos.startswith('VB'):  # Verb
            verb_count += 1
        elif pos.startswith('NN'):  # Noun
            noun_count += 1
        elif pos.startswith('PRP'):  # Pronoun
            pronoun_count += 1
        elif pos.startswith('JJ'):  # Adjective
            adjective_count += 1

    # Create a dictionary of the counts
    pos_counts = {
        'Verbs': verb_count,
        'Nouns': noun_count,
        'Pronouns': pronoun_count,
        'Adjectives': adjective_count
    }

    return pos_counts

text1 = "The police caught the robber and got the robbed items."
result1 = count_pos_tags(text1)
print("text1")
print(result1)

text2 = "He Jumped along the rainy street, feeling the cold raindrops on his body."
result2 = count_pos_tags(text2)
print("text2")
print(result2)

text3 = "The greatest player in the world, one and only Virat Kohli."
result3 = count_pos_tags(text3)
print("text3")
print(result3)
```
