```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("/content/sample_data/data.csv")
data1.head()
```

```python
data1 = data1.sample(frac=1).reset_index(drop=True)
data1.head()
```

```python
data2 = pd.read_csv("/content/sample_data/data_2genre.csv")
data2.head()
```

```python
data2['label'].loc[data2['label'] == 1] = 'pop'
data2['label'].loc[data2['label'] == 2] = 'classical'
data2.head()
```

```python
data1["label"].unique()
```

```python
data1.columns
```

```python
col = ['mfcc1', 'mfcc2', 'mfcc3', 'mfcc4', 'mfcc5',
       'mfcc6', 'mfcc7', 'mfcc8', 'mfcc9', 'mfcc10', 'mfcc11', 'mfcc12',
       'mfcc13', 'mfcc14', 'mfcc15', 'mfcc16', 'mfcc17', 'mfcc18', 'mfcc19',
       'mfcc20']
X = data1[col]
y = data1['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=10)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
acc
```

```python
y_pred2 = model.predict(data2[col])
y_pred2
```

```python
print("Accuracy of Prediction is:", accuracy_score(y_pred2, data2['label']))
```

```python
v_data = data1.iloc[:,1:29]
v_data.info()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
data_pca = pca.fit_transform(v_data)
```

```python
pca.explained_variance_ratio_
```

```python
new_data = pd.DataFrame(data = data_pca, columns = ['principal component 1', 'principal component 2','principal component 3' ])
```

```python
final_data = pd.concat([new_data, data1['label']], axis = 1)
final_data.head()
```

~ Display of the genres of music :-

```python
import plotly.express as px
fig = px.scatter(data_pca, x = 0, y = 1, color = final_data['label'])
fig.show()
```
